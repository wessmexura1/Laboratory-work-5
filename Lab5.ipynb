```python
import numpy as np
import time
import random
```

```python
def create_diag_numpy(n):
    
    matrix = np.zeros((n, n), dtype=int)
    
    diagonal_values = np.arange(1, n + 1)
    
    np.fill_diagonal(matrix, diagonal_values)
    
    return matrix
```

```python
def create_diag_python(n):
    matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == j:
                row.append(i + 1)
            else:
                row.append(0)
        matrix.append(row)
    return matrix
```

```python
def task1():
    n = int(input("Введіть розмір n: "))

    start_time = time.perf_counter()
    matrix_numpy = create_diag_numpy(n)
    end_time = time.perf_counter()
    numpy_time = end_time - start_time

    start_time = time.perf_counter()
    matrix_python = create_diag_python(n)
    end_time = time.perf_counter()
    python_time = end_time - start_time

    print("Матриця (NumPy):")
    print(matrix_numpy)

    print("Матриця (списки/цикли):")
    for row in matrix_python:
        print(row)

    print(f"Час NumPy: {numpy_time:.6f} с")
    print(f"Час списки/цикли: {python_time:.6f} с")

    if numpy_time < python_time:
        print("Варіант з NumPy працює швидше.")
    elif numpy_time > python_time:
        print("Варіант зі звичайними циклами працює швидше")
    else:
        print("Час виконання обох варіантів приблизно однаковий.")
```

```python
def create_matrix_numpy(n, m, r, c):
    matrix = np.ones((n, m), dtype=int)
    matrix[r - 1::r, :] = 0
    matrix[:, c - 1::c] = 0
    return matrix
```

```python
def create_matrix_python(n, m, r, c):
    matrix = []
    for i in range(1, n + 1):
        row = []
        for j in range(1, m + 1):
            if i % r == 0 or j % c == 0:
                row.append(0)
            else:
                row.append(1)
        matrix.append(row)
    return matrix
```

```python
def frame_numpy(n):
    matrix = np.ones((n, n), dtype=int)
    matrix[0, :] = 0
    matrix[-1, :] = 0
    matrix[:, 0] = 0
    matrix[:, -1] = 0
    return matrix
```

```python
def frame_python(n):
    matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == 0 or j == 0 or i == n - 1 or j == n - 1:
                row.append(0)
            else:
                row.append(1)
        matrix.append(row)
    return matrix
```

```python
def cut_vector_numpy(n):
    vec = np.arange(0, n + 1, dtype=int)
    lower = n / 4
    upper = 3 * n / 4
    mask = (vec > lower) & (vec < upper)
    vec[mask] = 0
    return vec
```

```python
def cut_vector_python(n):
    vec = list(range(0, n + 1))
    lower = n / 4
    upper = 3 * n / 4
    for i in range(len(vec)):
        if vec[i] > lower and vec[i] < upper:
            vec[i] = 0
    return vec
```

```python
def task16():
    n = int(input("Введіть n: "))

    start = time.perf_counter()
    vec_numpy = cut_vector_numpy(n)
    end = time.perf_counter()
    numpy_time = end - start

    start = time.perf_counter()
    vec_python = cut_vector_python(n)
    end = time.perf_counter()
    python_time = end - start

    print("Вектор (NumPy):")
    print(vec_numpy)

    print("Вектор (списки/цикли):")
    print(vec_python)

    print(f"Час NumPy: {numpy_time:.6f} с")
    print(f"Час списки/цикли: {python_time:.6f} с")

    if numpy_time < python_time:
        print("Варіант з NumPy працює швидше.")
    elif numpy_time > python_time:
        print("Варіант зі звичайними циклами працює швидше")
    else:
        print("Час виконання обох варіантів приблизно однаковий.")
```

## Завдання №2 - Розв’язати систему за допомогою формул Крамера і виконати перевірки

\begin{cases}
x_1 + 2x_2 + 3x_3 + 4x_4 = 5,\\
2x_1 + x_2 + 2x_3 + 3x_4 = 1,\\
3x_1 + 2x_2 + x_3 + 2x_4 = 1,\\
4x_1 + 3x_2 + 2x_3 + x_4 = -5
\end{cases}

```python
def solve_linear_system_cramers_rule(coefficient_matrix, constant_vector):
    determinant_main_matrix = np.linalg.det(coefficient_matrix)

    if abs(determinant_main_matrix) < 1e-12:
        raise ValueError("Визначник головної матриці близький до нуля, система може не мати єдиного розв'язку.")

    number_of_unknowns = coefficient_matrix.shape[0]
    solution_vector = np.zeros(number_of_unknowns, dtype=float)

    for column_index in range(number_of_unknowns):

        replaced_matrix = coefficient_matrix.copy()
        replaced_matrix[:, column_index] = constant_vector

        determinant_replaced_matrix = np.linalg.det(replaced_matrix)
        solution_vector[column_index] = determinant_replaced_matrix / determinant_main_matrix

    return solution_vector
```

```python
def task_linear_system():

    coefficient_matrix = np.array([[1, 2, 3, 4],
                                   [2, 1, 2, 3],
                                   [3, 2, 1, 2],
                                   [4, 3, 2, 1]], dtype=float)

    constant_vector = np.array([5, 1, 1, -5], dtype=float)

    print("Матриця коефіцієнтів (A):")
    print(coefficient_matrix)
    print("Вектор вільних членів (b):")
    print(constant_vector)

    cramers_solution = solve_linear_system_cramers_rule(coefficient_matrix, constant_vector)
    print("\nРозв'язок за формулами Крамера:")
    print(cramers_solution)

    left_side_values = coefficient_matrix @ cramers_solution
    print("\nПеревірка через матричне множення A @ x:")
    print("A @ x =", left_side_values)
    print("b =", constant_vector)
    print("Чи співпадає A @ x з b? ->", np.allclose(left_side_values, constant_vector))

    inverse_matrix = np.linalg.inv(coefficient_matrix)
    inverse_solution = inverse_matrix @ constant_vector
    print("\nРозв'язок через обернену матрицю A^(-1) * b:")
    print(inverse_solution)

    solve_function_solution = np.linalg.solve(coefficient_matrix, constant_vector)
    print("\nРозв'язок через numpy.linalg.solve:")
    print(solve_function_solution)

    print("\nПорівняння розв'язків :") #(numpy.allclose)
    print("Крамер vs inverse:", np.allclose(cramers_solution, inverse_solution))
    print("Крамер vs solve  :", np.allclose(cramers_solution, solve_function_solution))
    print("inverse vs solve :", np.allclose(inverse_solution, solve_function_solution))
```

## Завдання №3 - Обчислити значення матричного виразу

$$2(A-B)(A^{2}+B)B$$


A=\begin{pmatrix}
5 & 1 & 7\\
-10 & -2 & 1\\
0 & 1 & 2
\end{pmatrix}
B=\begin{pmatrix}
2 & 4 & 1\\
3 & 1 & 0\\
7 & 2 & 1
\end{pmatrix}

```python
def calculate_matrix_expression_numpy():

    matrix_A = np.array([[5, 1, 7],
                         [-10, -2, 1],
                         [0, 1, 2]], dtype=float)

    matrix_B = np.array([[2, 4, 1],
                         [3, 1, 0],
                         [7, 2, 1]], dtype=float)

    matrix_A_squared = matrix_A @ matrix_A      # A^2
    matrix_difference = matrix_A - matrix_B     # A - B
    matrix_sum = matrix_A_squared + matrix_B    # A^2 + B

    result_matrix = 2 * (matrix_difference @ matrix_sum)

    return result_matrix
```

```python
def subtract_two_matrices(first_matrix, second_matrix):
    rows_count = len(first_matrix)
    columns_count = len(first_matrix[0])
    result_matrix = []

    for row_index in range(rows_count):
        new_row = []
        for column_index in range(columns_count):
            new_value = first_matrix[row_index][column_index] - second_matrix[row_index][column_index]
            new_row.append(new_value)
        result_matrix.append(new_row)

    return result_matrix


def add_two_matrices(first_matrix, second_matrix):
    rows_count = len(first_matrix)
    columns_count = len(first_matrix[0])
    result_matrix = []

    for row_index in range(rows_count):
        new_row = []
        for column_index in range(columns_count):
            new_value = first_matrix[row_index][column_index] + second_matrix[row_index][column_index]
            new_row.append(new_value)
        result_matrix.append(new_row)

    return result_matrix


def multiply_two_matrices(first_matrix, second_matrix):
    rows_count = len(first_matrix)
    common_size = len(second_matrix)
    columns_count = len(second_matrix[0])
    result_matrix = []

    for row_index in range(rows_count):
        new_row = []
        for column_index in range(columns_count):
            cell_sum = 0
            for common_index in range(common_size):
                cell_sum += first_matrix[row_index][common_index] * second_matrix[common_index][column_index]
            new_row.append(cell_sum)
        result_matrix.append(new_row)

    return result_matrix


def multiply_matrix_by_number(number_value, matrix):
    result_matrix = []
    for row in matrix:
        new_row = []
        for element in row:
            new_row.append(number_value * element)
        result_matrix.append(new_row)
    return result_matrix


def calculate_matrix_expression_with_loops():

    matrix_A = [[5, 1, 7],
                [-10, -2, 1],
                [0, 1, 2]]

    matrix_B = [[2, 4, 1],
                [3, 1, 0],
                [7, 2, 1]]

    matrix_A_squared = multiply_two_matrices(matrix_A, matrix_A)
    matrix_difference = subtract_two_matrices(matrix_A, matrix_B)
    matrix_sum = add_two_matrices(matrix_A_squared, matrix_B)

    product_matrix = multiply_two_matrices(matrix_difference, matrix_sum)
    final_result_matrix = multiply_matrix_by_number(2, product_matrix)

    return final_result_matrix
```

```python
def task_matrix_expression():

    start_time = time.perf_counter()
    numpy_result = calculate_matrix_expression_numpy()
    end_time = time.perf_counter()
    numpy_execution_time = end_time - start_time

    start_time = time.perf_counter()
    loops_result = calculate_matrix_expression_with_loops()
    end_time = time.perf_counter()
    loops_execution_time = end_time - start_time

    print("\nРезультат (NumPy):")
    print(numpy_result)

    print("\nРезультат (списки та цикли):")
    for row in loops_result:
        print(row)

    loops_result_as_array = np.array(loops_result, dtype=float)
    print("\nЧи однакові результати? ->",
          np.allclose(numpy_result, loops_result_as_array))

    print(f"\nЧас виконання NumPy: {numpy_execution_time:.6f} с")
    print(f"Час виконання цикли : {loops_execution_time:.6f} с")

    if numpy_execution_time < loops_execution_time:
        print("Варіант з NumPy працює швидше.")
    elif numpy_execution_time > loops_execution_time:
        print("На цих розмірах матриць швидше спрацювали звичайні цикли.")
    else:
        print("Час виконання двох варіантів приблизно однаковий.")
```
